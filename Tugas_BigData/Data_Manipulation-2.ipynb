```python
import pandas as pd
# File CSV
df_csv = pd.read_csv('https://raw.githubusercontent.com/rahmadsa/dataset/main/sample_csv.csv')
# File TSV
df_tsv = pd.read_table('https://raw.githubusercontent.com/rahmadsa/dataset/main/sample_tsv.tsv', sep='\t')
```

```python
df_12231949 = pd.concat([df_csv, df_tsv], ignore_index=True)
print("Jumlah data CSV :", len(df_csv))
print("Jumlah data TSV :", len(df_tsv))
print("Jumlah data gabungan :", len(df_12231949))
```
